```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\asus\Desktop\BostonHousing.csv")
df
```

```python
X = df.drop(columns = "medv")
y = df["medv"]
```

*  medv is median value of house price

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=133)
```

```python
X_train
```

* Model Building with Tensor flow

```python
# Define the network architecture (Input, Output, Hidden Layer, Nodes)
# Model Compilation Define the type of function(Loss Function)
# Model Train - Pass Dataset, Iteration, Test Data
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from keras.optimizers import *
```

```python
model = Sequential()
model.add(Input(13))
model.add(Dense(5, activation="relu"))
# model.add(Dense(5, activation="sigmoid"))

model.add(Dense(1)) # Output
```

```python

model.compile(optimizer = Adam(learning_rate = 0.01), loss = "mean_squared_error")  # Algorithm that we require for backpropogation
```

```python
# Train Method
model.fit(x = X_train, y = y_train, epochs= 100, validation_data = (X_test, y_test))
```

```python
history = model.history.history
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_loss = history["loss"]
val_loss = history["val_loss"]
plt.plot(train_loss)
plt.plot(val_loss)
plt.ylim(0, 500)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss Curve(MSE)")
plt.grid()
plt.show()
```

```python
model.weights
```

```python
y_pred = model.predict(X_test).flatten()
```

```python
pd.DataFrame({"Actual": y_test, "Preds":y_pred }).head(20)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
