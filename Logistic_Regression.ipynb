## Converting the object type into 1 and 0 and type to string

```python
data["y"] = np.where(data["y"] == "yes", 1, 0)
```

```python
data["y"] = data["y"].astype(str)
```

## Checking the null values



```python
data.isnull().sum().sum()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
data.describe()
```

```python
data.info()
```

```python
data["y"].value_counts()
```

## Find the IQR and slicing the dataset

```python
data = data[~(((data < (Q1 - 1.5 * IQR)) |( data > (Q3 + 1.5 * IQR))).any(axis=1))]
```

```python
data
```

```python
data.shape
```

```python
col_list = list(data.columns)
col_list
```

## Removing dependent column from list of independent columns

```python
col_list.remove("y")
```

```python
col_list
```

## Replacing the null values with mean amd mode

```python
for col in col_list:
    if data[col].dtypes == "object":
        data[col] = data[col].fillna(data[col].mode()[0])
    else:
        data[col] = data[col].fillna(data[col].mean())
```

```python
data.isna().sum()
```

```python
data["marital"].value_counts()
```

```python
data.info()
```

## Will apply label encoding since most of the data is in object format

```python
bins = [0, 1, 5, 10, 25, 50, 100]
data["age"] = pd.cut(data["age"], bins)
data["age"] = data.age.astype(str)

```

```python
data.tail(10)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for i in col_list:
    data[i] = labelencoder.fit_transform(data[i])
```

```python
data
```

## Finding the Variance Inflation Factor and removing the colummns which have VIF more than 5

###  Variance Inflation Factor


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in data.columns:
    if (( data[col].dtype != "object") & (col != "y")):
        col_list.append(col)
    
X = data[col_list]
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range (len(X.columns))]
vif_data
```

```python
data = data.drop(["age", "pdays", "previous", "poutcome", "emp.var.rate", "cons.price.idx", "euribor3m", "nr.employed"], axis = 1)
```
