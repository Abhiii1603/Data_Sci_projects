```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

%matplotlib inline
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer["target"]
```

```python
cancer.DESCR
```

```python
df = pd.DataFrame(cancer["data"], columns = cancer["feature_names"])
```

```python
df
```

```python
df.sample(10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
x_pca
```

```python
scaled_data.shape, x_pca.shape
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer["target"])

plt.xlabel("First PCA")

plt.ylabel("Second PCA")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(x_pca[:, 0], x_pca[:, 1], x_pca[:, 2], c=cancer["target"], cmap='viridis')


ax.set_xlabel("First PCA")
ax.set_ylabel("Second PCA")
ax.set_zlabel("Third PCA")

# Adding colorbar
colorbar = plt.colorbar(scatter)
colorbar.set_label("Cancer Target")

plt.show()
```
